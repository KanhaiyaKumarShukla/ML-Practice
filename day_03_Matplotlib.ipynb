````python
# dataset is big so we have used 25 rows random sample, with its name, runs, avg, strike_rate
sample_df=df.head(100).sample(25, random_state=5)[['name', 'Runs', 'avg', 'Strike_rate']]
sample_df.head()
````

* s: using is parameter we can vary the size of marker according to any numeric data. i have divided the 'runs' column by 5 because it was causing much bigger circle.

* we can add text to the marker to identify. Parameters:
     * x, y : the cordinate of the text. it may be the cordinate of marker.
     * s: text to add to the marker.
     * fontdict: we can modify size, color, etc using dictionary.

````python
plt.figure(figsize=(20, 14))
plt.scatter(sample_df['avg'], sample_df['Strike_rate'], s=sample_df['Runs']/5)

for i in range(len(sample_df)):
  plt.text(sample_df['avg'].values[i], sample_df['Strike_rate'].values[i], sample_df['name'].values[i], fontdict={'size':15})
````

* `axhline()` function is used to draw a horizontal line across the axes.


```
plt.axhline(y=0, color='r', linestyle='-')
```
y: y-coordinate
 * `axvline()` is used to draw a vertical line across the axes.


```
plt.axvline(y=0, color='r', linestyle='-')
```




````python
plt.figure(figsize=(20, 14))
plt.scatter(sample_df['avg'], sample_df['Strike_rate'], s=sample_df['Runs']/5)

plt.axhline(95, color='red')
plt.axvline(40, color='green')


for i in range(len(sample_df)):
  plt.text(sample_df['avg'].values[i], sample_df['Strike_rate'].values[i], sample_df['name'].values[i], fontdict={'size':15})
````

There is another way to show graph:
### subplots: we can use this to  arranging multiple plots in a grid layout.


```
plt.subplots(nrows=1, ncols=1, sharex=False, sharey=False)
```
Parameters:
* nrows:number of rows of subplots.
* ncols: number of columns of subplots.
* sharex: If True, the x-axis will be shared among all subplots.
* sharey: If True, the y-axis will be shared among all subplots.
* figsize: Set the figure size, e.g., figsize=(width, height).


subplot returns:
* fig: The figure object.
* ax: An array of axes objects.

````python
data = {
    'Batsman': ['Virat Kohli', 'AB de Villiers', 'Joe Root', 'Steve Smith', 'Kane Williamson', 'David Warner'],
    '2015': [843, 1023, 800, 952, 1054, 897],
    '2016': [973, 876, 850, 914, 932, 748],
    '2017': [896, 745, 945, 1082, 870, 826]
}

# Creating the DataFrame
df = pd.DataFrame(data)
df
````

````python
# to show bar chart of all season for all batsman:
'''
we have to arrange the bar for each year so we have first calculated df shape (6) and arange will give [0, 1,....5]
and then displace it by 0.2 distance.
adjusted width.
'''
plt.bar(np.arange(df.shape[0])-0.2, df['2015'], width=0.2)
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.2)
plt.bar(np.arange(df.shape[0])+0.2, df['2017'], width=0.2)

# now the x axis is labeled 0, 1, 2, 3... for this we use xticks(old_lebel, new_lebel )
plt.xticks(np.arange(df.shape[0]), df['Batsman'])
````

````python
# if lebel is too big as above it will overlap. for this we use rotation parameter in plt.xticks().
plt.bar(np.arange(df.shape[0])-0.2, df['2015'], width=0.2)
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.2)
plt.bar(np.arange(df.shape[0])+0.2, df['2017'], width=0.2)

# rotation = 'vertical'/ 'horizontal' / number like 40, 20, 60, ....
plt.xticks(np.arange(df.shape[0]), df['Batsman'], rotation=70)
````

````python
# stacked bar char: in bottom para, we pass the addition of all the series which will be bellow it.

plt.bar(df['Batsman'], df['2017'], label='2017')
plt.bar(df['Batsman'], df['2016'], bottom=df['2017'], label='2016')
plt.bar(df['Batsman'], df['2015'], bottom=(df['2017']+ df['2016']), label='2015')
plt.legend()
````

### 3D Surface Plots:

````python
x=np.linspace(-10, 10, 100)
y=np.linspace(-10, 10, 100)
xx, yy=np.meshgrid(x, y)  # create 2d grid of x, y
print(xx.shape, yy.shape)
````

````python
z=xx**2 + yy**2    # equation is z=x^2 + y^2

fig=plt.figure()
ax=fig.add_subplot(projection='3d')
ax.plot_surface(xx, yy, z)
````

#### Contour Plots

````python
ax=plt.subplot()
ax.contour(xx, yy, z)
````

````python
ax=plt.subplot()
ax.contourf(xx, yy, z)  # filled contour
````

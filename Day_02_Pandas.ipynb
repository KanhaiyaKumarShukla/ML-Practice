### rank(): to assign ranks to elements within a DataFrame or Series. It can rank the elements in ascending or descending order, and it has options to deal with ties (when two elements have the same value).

### parameters:
* ascending: If True (default), the lowest values are assigned the lowest ranks. If False, the largest values get the lowest ranks.
* method: Defines how to rank if there are ties. Possible values are:

    * 'average': (default) Assigns the average rank of the tied group.
    * 'min': Assigns the minimum rank to all tied values.
    * 'max': Assigns the maximum rank to all tied values.
    * 'first': Assigns ranks in the order the values appear in the data.
    * 'dense': Like min, but the rank values are consecutive.

````python
team_wins = data['winner'].value_counts().reset_index()   # reset_index() will convert the series into DF.
team_wins.columns = ['team', 'wins']    # rename the columns
team_wins['rank'] = team_wins['wins'].rank(method='min', ascending=False)
team_wins
````

````python
def players_run_against_teams(players, count):
  return deliveries[deliveries['batsman']==players].groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(count)

players_run_against_teams('S Dhawan', 8)
````

````python
def most_runs_scored(count):
  return deliveries.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(count)

most_runs_scored(10)
````

````python
def most_runs_scored_against_bowlers(batsman, count):
  return deliveries[deliveries['batsman']==batsman].groupby('bowler')['batsman_runs'].sum().sort_values(ascending=False).head(count)

most_runs_scored_against_bowlers('V Kohli', 10)
````

apply() function to apply custom function on groupby object.

````python
# find number of companies starting with 'A' for each group:

def foo(group):
  return group['Company'].str.startswith('A').sum()

industries.apply(foo)
````

````python
# find the rank of each companies in group according to 'investors count'.
def rank(group):
  group['rank']=group['Investors Count'].rank(ascending=False)
  return group

industries.apply(rank)
````

````python
# find normalized investors count group wise. search normalization formula
def foo(group):
  minm, maxm=group['Investors Count'].min(), group['Investors Count'].max()
  group['normal_rank']=(group['Investors Count']-minm)/(maxm-minm) if maxm!=minm else 0
  return group

industries.apply(foo)
````

### dropna(): used to remove missing data (i.e., rows or columns with NaN values).


```
DataFrame.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)

```

* axis: Specifies whether to drop rows or columns.

    * axis=0 (default) drops rows with missing values.
    * axis=1 drops columns with missing values.
* how: Determines whether to drop rows/columns based on the presence of any or all missing values.

    * 'any' (default): Drops rows/columns if any value is missing.
    * 'all': Drops rows/columns only if all values are missing.

* subset: Specifies which rows/columns to consider when looking for missing values (default considers all).

* inplace: If True, modifies the DataFrame in place without returning a new one. If False (default), returns a new DataFrame.



````python
deliveries.dropna().shape   # by default how='any', so it drop all rows which have at least 1 non-NA values.
````

### fillna():  used to fill missing (NaN) values with a specific value or a strategy such as forward/backward fill or filling with the mean/median of a column.


```
DataFrame.fillna(value=None, method=None, axis=None, inplace=False, limit=None, downcast=None)

```
* value: The value used to replace missing data.
* method: Specifies how to fill missing values:

    * 'ffill': Forward fill. It replace with previous valid value.
    * 'bfill': Backward fill. It replace next valid value.

````python
def foo(group):
    group['sixes'] = (group['batsman_runs'] == 6).sum()
    group['fours'] = (group['batsman_runs'] == 4).sum()
    group['total_runs_per_team'] = group['batsman_runs'].sum()
    return group

# Filter data for 'V Kohli'
kohli_data = ipl[ipl['batsman'] == 'V Kohli']

# Group by bowling team and apply the function to calculate sixes, fours, total runs
kohli_stats = kohli_data.groupby('bowling_team').apply(foo).drop_duplicates(['sixes', 'fours', 'total_runs_per_team'])

# Reset index for kohli_stats but do not insert the existing index (bowling_team) as a column
kohli_stats = kohli_stats.reset_index(drop=True)
# Calculate dismissals (how many times Kohli was out against each bowling team)
dismissals = kohli_data[kohli_data['player_dismissed'] == 'V Kohli'].groupby('bowling_team')['player_dismissed'].count().reset_index().rename(columns={'player_dismissed': 'dismissals'}) #  renaming the player_dismissed column to dismissals.

# Add dismissals to the kohli_stats dataframe
kohli_stats = kohli_stats.merge(dismissals, on='bowling_team', how='left')

# Calculate the average runs per team (total runs / dismissals)
kohli_stats['avg'] = kohli_stats['total_runs_per_team'] / kohli_stats['dismissals']

# Display the result
kohli_stats[['bowling_team', 'sixes', 'fours', 'total_runs_per_team', 'dismissals', 'avg']]
````

another ways of solving this:
`apply()` on a grouped object can be slower for larger datasets because it essentially loops over groups.
use `agg()` for better performance

````python
kohli_data = ipl[ipl['batsman'] == 'V Kohli']
kohli_stats = kohli_data.groupby('bowling_team').agg(
    sixes=('batsman_runs', lambda x: (x == 6).sum()),  # Count sixes
    fours=('batsman_runs', lambda x: (x == 4).sum()),  # Count fours
    total_runs_per_team=('batsman_runs', 'sum')  # Sum total runs
).reset_index()

dismissals = kohli_data[kohli_data['player_dismissed'] == 'V Kohli'] \
    .groupby('bowling_team')['player_dismissed'].count() \
    .reset_index().rename(columns={'player_dismissed': 'dismissals'})

kohli_stats = kohli_stats.merge(dismissals, on='bowling_team', how='left')

kohli_stats['avg'] = kohli_stats['total_runs_per_team'] / kohli_stats['dismissals']

kohli_stats[['bowling_team', 'sixes', 'fours', 'total_runs_per_team', 'dismissals', 'avg']]
````
